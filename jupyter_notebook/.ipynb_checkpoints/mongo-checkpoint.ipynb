```python
from pymongo import MongoClient
import pandas as pd

# Connect to MongoDB server
client = pymongo.MongoClient("mongodb://root:pass12345@113.53.253.56:27017/")  # Replace with your MongoDB server URL

# Create or select a database
db = client["water_balance_db"]

# Create or select a collection within the database
collection = db["precip_onemap_khs_59TB"]

# Identify Columns to Drop
cursor = collection.find()

# Initialize a set to store the names of columns with NaN values
columns_to_drop = set()

# # Iterate through documents to find columns with NaN values
# for doc in cursor:
#     for key, value in doc.items():
#         if pd.isna(value):
#             columns_to_drop.add(key)

# Drop Columns with NaN Values
for doc in collection.find():
    updated_doc = {key: value for key, value in doc.items() if key not in columns_to_drop}
    collection.update_one({'_id': doc['_id']}, {'$set': updated_doc})

# Print the names of columns that were dropped
print(f"Columns with NaN values dropped: {', '.join(columns_to_drop)}")

# Close the MongoDB connection
client.close()
```

```python
from pymongo import MongoClient

# Connect to MongoDB server
client = pymongo.MongoClient("mongodb://root:pass12345@113.53.253.56:27017/")  # Replace with your MongoDB server URL

# Create or select a database
db = client["water_balance_db"]

# Create or select a collection within the database
collection = db["precip_onemap_khs_59TB"]

# Create an aggregation pipeline to identify columns with NaN values
pipeline = [
    {
        "$project": {
            "_id": 0,  # Exclude the _id field
            "columns": {
                "$objectToArray": "$$ROOT"  # Convert each document to an array of key-value pairs
            }
        }
    },
    {
        "$unwind": "$columns"  # Unwind the array of key-value pairs
    },
    {
        "$match": {
            "columns.v": {"$type": 10}  # Match where the value is NaN (null)
        }
    },
    {
        "$group": {
            "_id": "$columns.k",  # Group by column names
            "count": {"$sum": 1}   # Count occurrences of NaN in each column
        }
    },
    {
        "$match": {
            "count": {"$eq": {"$size": db.collection.estimated_document_count()}}
            # Only keep columns where the count of NaN matches the number of documents
        }
    }
]

# Execute the aggregation pipeline to identify columns with all NaN values
columns_to_drop = [doc["_id"] for doc in collection.aggregate(pipeline)]

# Drop identified columns from each document in the collection
for column in columns_to_drop:
    collection.update_many({}, {"$unset": {column: 1}})

# Print the names of columns that were dropped
print(f"Columns with all NaN values dropped: {', '.join(columns_to_drop)}")

# Close the MongoDB connection
client.close()
```

```python
from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient('mongodb://localhost:27017/')

# Specify the database and collection
db = client['your_database']
collection = db['your_collection']

# Find all documents
documents = list(collection.find())

# Function to remove fields with NaN values
def remove_nan_fields(doc):
    return {k: v for k, v in doc.items() if v != 'NaN'}

# Remove fields with NaN values from each document
filtered_documents = [remove_nan_fields(doc) for doc in documents]

# Print the filtered documents
for doc in filtered_documents:
    print(doc)

# Close the MongoDB connection
client.close()
```
